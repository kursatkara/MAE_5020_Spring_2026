## 3) NumPy arrays + indexing (5 minutes)
**Key rule:** Python uses **0-based indexing** and slices **exclude the end**.


```python
x = np.array([10, 20, 30, 40, 50], dtype=float)

print("x      =", x)
print("x[0]   =", x[0])     # first
print("x[-1]  =", x[-1])    # last
print("x[:3]  =", x[:3])    # first 3 elements: indices 0,1,2
print("x[2:]  =", x[2:])    # from index 2 to end
```

## 4) Quick vectorized math (2 minutes)


```python
t = np.linspace(0, 2*np.pi, 200)
y = np.sin(t) + 0.2*np.cos(3*t)

print("t.shape:", t.shape, "y.shape:", y.shape)
print("mean(y) =", y.mean(), "std(y) =", y.std(ddof=1))
```

## 5) Plot (3 minutes)


```python
plt.figure()
plt.plot(t, y)
plt.xlabel("t")
plt.ylabel("y(t)")
plt.grid(True)
plt.show()
```

## 6) Load a tiny dataset (3 minutes)
We’ll simulate a small CSV and load it with NumPy.


```python
import io

csv_text = """t,V
0.0,10.0
0.5,12.2
1.0,14.1
1.5,13.7
2.0,15.5
2.5,16.0
3.0,15.2
"""

data = np.genfromtxt(io.StringIO(csv_text), delimiter=",", names=True)
t_data = data["t"]
V = data["V"]

print("t:", t_data)
print("V:", V)
print("mean(V) =", V.mean(), "std(V) =", V.std(ddof=1))

plt.figure()
plt.plot(t_data, V, "o-")
plt.xlabel("t (s)")
plt.ylabel("V (m/s)")
plt.grid(True)
plt.show()
```
