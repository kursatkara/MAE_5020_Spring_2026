```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.misc

plt.rcParams['figure.figsize'] = [6,6]
plt.rcParams.update({'font.size': 12})

n = 1000
q = int(n/4)
# Black
X = np.zeros((n,n))
# White
X[(q-1):(3*q),(q-1):(3*q)] = 1

plt.imshow(X)
plt.set_cmap('gray')
plt.axis('off')
plt.title('Unrotated Matrix')
plt.show()
```

```python
# SVD of Unrotated Matrix

U, S, VT = np.linalg.svd(X,full_matrices=0)

plt.semilogy(S,'-o',color='k')
plt.ylim((10**(-16),10**(4)+1))
plt.yticks(np.power(10,np.arange(-16,4,4, dtype=float)))
plt.xticks(np.arange(0,1000,250))
plt.grid()
plt.title('Unrotated Matrix: Spectrum')
plt.show()
```

```python
print(S[0:4])
```

```python
# Extract the first column from U
y_values = U[:,0]

# Create x-values as indices from 0 to the length of the column - 1
x_values = np.arange(len(y_values))

# Create the plot
plt.plot(y_values, x_values, marker='o')  # 'o' denotes circle markers

# Add title and labels (optional)
plt.title('Plot of the First Column of U_rot')
plt.ylabel('Index')
plt.xlabel('Value')

# Show the plot
plt.show()
```

```python
# Extract the first row of VT
y_values = VT[0,:]

# Create x-values as indices from 0 to the length of the column - 1
x_values = np.arange(len(y_values))

# Create the plot
plt.plot(x_values, y_values, marker='o')  # 'o' denotes circle markers

# Add title and labels (optional)
plt.title('Plot of the First Row of VT')
plt.xlabel('Index')
plt.ylabel('Value')

# Show the plot
plt.show()
```

```python
# Rotated Matrix
import skimage.transform

X_rot = skimage.transform.rotate(X,35)
X_rot[np.nonzero(X_rot)] = 1


plt.imshow(X_rot)
plt.set_cmap('gray')
plt.axis('off')
plt.title('Rotated Matrix')
plt.show()
```

```python
# SVD of the Rotated Matrix

U_rot, S_rot, VT_rot = np.linalg.svd(X_rot,full_matrices=0)

plt.semilogy(S_rot,'-o',color='k')
plt.ylim((10**(-16),10**(4)+1))
plt.yticks(np.power(10,np.arange(-16,4,4, dtype=float)))
plt.xticks(np.arange(0,1000,250))
plt.grid()
plt.title('Rotated Matrix: Spectrum')
plt.show()
```

```python
print(S_rot[0:4])
```

```python
# Extract the first column from U
y_values = U_rot[:,0]

# Create x-values as indices from 0 to the length of the column - 1
x_values = np.arange(len(y_values))

# Create the plot
plt.plot(y_values, x_values, marker='o')  # 'o' denotes circle markers

# Add title and labels (optional)
plt.title('Plot of the First Column of U')
plt.ylabel('Index')
plt.xlabel('Value')

# Show the plot
plt.show()
```

```python
# Extract the first row of VT
y_values = VT_rot[0,:]

# Create x-values as indices from 0 to the length of the column - 1
x_values = np.arange(len(y_values))

# Create the plot
plt.plot(x_values, y_values, marker='o')  # 'o' denotes circle markers

# Add title and labels (optional)
plt.title('Plot of the First Row of VT')
plt.xlabel('Index')
plt.ylabel('Value')

# Show the plot
plt.show()
```
